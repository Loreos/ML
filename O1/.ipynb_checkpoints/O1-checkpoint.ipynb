## Qa) The $\theta$ parameters and the $R^2$ Score

### $R^2$ Score / Coefficient of determination
The score of the model is extracted as

````python
R2lin = model.score(X, y)
print(f"R2 value of the linear regression model is {round(R2lin,4)}")
````

The $R^2$ score in linear regression is a comparison of the regression to a horizontal line, to see how much better the regression fits the data. </br>

1 is the best value of $R^2$. A score closer to 1 is better. If the score is 0, it means that the model fits as well as a horizontal line, while negatvie values mean the model fit worse than a horizontal line.
$R^2$ can at most have a value of 1, but can go negative infinitely.  </br>

$R^2$ is not a loss or objective function, but a function that measures the fitness of the model.

## Qe Extend the class with some public and private functions and member variables
Private functions and memebr variables in Python are represented with double underscores, `__` , as the first letters in the name. The following example shows a class with a private and a public variable, a private and a public method, and what happens when trying to call them.

````python
# class with a private function using "__"
class MyClass:
    var    = "blah"
    __var2 = "bleh"
    
    def myfun_public(self):
        print("This is a public function")
    
    def __myfun_private(self):
        print("This is a private function")

# This will display the message from the public function
c1 = MyClass()
c1.myfun_public()

# This will try to call the private function, but fail, and give an AttributeError.
try:
    c1.__myfun_private()
except AttributeError:
    print("Error: AttributeError. Cannot call the private function.")
    
#This will call the public variable, thend the private:
print("This is the public variable: " + c1.var)

try:
    print("This is the private variable: " + c1.__var2)
except AttributeError:
    print("Error: AttributeError. Cannot access the private variable.")
````

#### Self
`self` always refers to the particular instance of a class.

If we dont include `self` as the first argument for a method in a class, the method will not be available through an instance of a class, but only through the class. </br>
The following example shows what happens when a method in a class does NOT have `self` as the first argument. It results in a TypeError when trying to call it using the instance, since self refers to the instance. It can, however, be called directly from the class. 

````python
#For a class in which a method does NOT have self as the first argument, the method can only be used through the class, not the instance.
class MyClass:
    def myfun():
        print("hi")

a = MyClass() #Instance of the class

#This will print "hi"
MyClass.myfun()

#This cannot print "hi"
a.myfun()
    
````

The next example shows what happens what a method in a class DOES have self as the first argument. Calling it from the instance works, but now calling it from the class results in a TypeError, since the class does not have an instance to pass to the method, as it needs due to the `self` argument.

````python
#For a class in which a method DOES have self as the first argument, the method can only be used through the instance, not the class.
class MyClass:
    def myfun(self):
        print("hi")

a = MyClass() #Instance of the class

#This will print "hi"
a.myfun()

#This will not print "hi"
MyClass.myfun()
````

## Qb Implement the $\norm{1}$ and $\norm{2}$ norms for vectors in python.

Setting up the norm functions:

````python
## L1 and L2 using explicit implementation

def L1(V):
    L1 = 0 #Start length at 0
    for x in V: #Sum all elements x
        L1 = L1 + (x**2)**0.5
    return L1  

def L2(V):
    L2 = 0 #Start length at 0
    for x in V: #Sum all elements of x^2
        L2 = L2 + ((x**2)**0.5)**2
    
    L2 = L2**0.5 #Take square root of sum
    return L2

#L2 using np.dot
def L2Dot(V):
    import numpy as np
    
    L2Dot = np.dot(V.T, V) #Find dot product
    
    L2Dot = np.sqrt(L2Dot) #Take square root of dot product
    
    return L2Dot
````

Testing the norm functions with code from the problem:

````python
# TEST vectors: here I test your implementation...calling your L1() and L2() functions
tx=np.array([1, 2, 3, -1])
ty=np.array([3,-1, 4,  1])

expected_d1=8.0
expected_d2=4.242640687119285

d1=L1(tx-ty)
d2=L2(tx-ty)

print(f"tx-ty={tx-ty}, d1-expected_d1={d1-expected_d1}, d2-expected_d2={d2-expected_d2}")

eps=1E-9 

import math

assert math.fabs(d1-expected_d1)<eps, "L1 dist seems to be wrong" 
assert math.fabs(d2-expected_d2)<eps, "L2 dist seems to be wrong" 

print("OK(part-1)")

#comment-in once your L2Dot fun is ready...
d2dot=L2Dot(tx-ty)
print("d2dot-expected_d2=",d2dot-expected_d2)
assert math.fabs(d2dot-expected_d2)<eps, "L2Ddot dist seem to be wrong" 
print("OK(part-2)")
````

## Qe Robust Code 

Error checking has been added to the RMSE, MAE, L2, L2Dot and L1 functions. They are tested further down.

````python


#RMSE function with error checking
def RMSE (ypred, y):
    
    assert ypred.ndim == 1, "First input is not a vector"
    assert y.ndim     == 1, "Second input is not a vector"
    assert len(ypred) == len(y), "Vectors do not have the same length."
    
    RMSE = (1/len(ypred)*L2(ypred - y)**2)**0.5
    return RMSE

#MAE function with error checking
def MAE(ypred, y):
    
    assert ypred.ndim == 1, "First input is not a vector"
    assert y.ndim     == 1, "Second input is not a vector"
    assert len(ypred) == len(y), "Vectors do not have the same length."
    
    MAE =  1/len(ypred)*L1(abs(ypred - y))
    return MAE

#L1 explicit
def L1(V):
    try:
        L1 = 0 #Start length at 0
        for x in V: #Sum all elements x
            L1 = L1 + (x**2)**0.5
        return L1  
    except:
        Print("input is not a vector")

#L2 explicit
def L2(V):
    try:
        L2 = 0 #Start length at 0

        for x in V: #Sum all elements of x^2
            L2 = L2 + ((x**2)**0.5)**2

        L2 = L2**0.5 #Take square root of sum
        return L2
    except:
        print("Input is not a vector")

#L2 using np.dot
def L2Dot(V):
    import numpy as np
    
    try:
        L2Dot = np.dot(V.T, V) #Find dot product
        L2Dot = np.sqrt(L2Dot) #Take square root of dot product
        return L2Dot
    except:
        print("Input is not a vector")
````

## Qa Implement the Accuracy function and test it on the MNIST data.
The following is a function which calculates the accuracy score based on the `y_pred` results and known `y_true`. It also checks if the data has any length, and if it does not have a length, returns the accuracy as NaN. This handles the case of the denominator being 0.

````python
def MyAccuracy(y_true, y_pred):
    # Using the formula a = (TP + TN)/N
    
    #Total number of datapoints
    N = len(y_true)
    
    #Initialize True Positive and True Negative counters
    TP = 0
    TN = 0
    
    if N == 0:
        print(f"Data is not valid. Length: {N}")
        a = float("NaN")
    else:
        for i in range(N):
            if y_true[i] == True and y_pred[i] == True: #True Positive
                TP = TP + 1
            elif y_true[i] == False and y_pred[i] == False: #True Negative
                TN = TN + 1
        a = (TP + TN)/N
    return a
````

Comparing the MyAccuracy function with the sklearn accuracy function on the SGD classifier and dummy classifier:

````python
from sklearn.metrics import accuracy_score

sgd_y_pred = sgd_classifier.predict(X_train)
dummy_pred = Dummy_5.predict(X_train)

# SGD Accuracy score with function and sklearn function 
MyA_SGD = MyAccuracy(y_train_5, sgd_y_pred)
a_SGD   = accuracy_score(y_train_5, sgd_y_pred)
print(f"SGD: \n My Accuracy: {MyA_SGD} \n Accuracy:    {a_SGD}")

# Dummy accuracy score with MyAccuracy Function and sklearn function
MyA_Dummy = MyAccuracy(y_train_5, dummy_pred)
a_Dummy   = accuracy_score(y_train_5, dummy_pred)
print(f"Dummy: \n My Accuracy: {MyA_Dummy} \n Accuracy:    {a_Dummy}")
````

## Qb Implement Precision, Recall and $F_1$-score and test it on the MNIST data for both the SGD and Dummy classifier models
To implement Precision, Recall and $F_1$-scores, the Precision and Recall functions are used to define the $F_1$ score, since it can be constructed from those.  
A function is created for each score type, as with the MyAccuracy function. 

````python
def MyPrecision(y_true, y_pred):
    # Using the formula a = (TP)/(TP+FP)
    
    #Total number of datapoints
    N = len(y_true)
    
    #Initialize True Positive and False Positive
    TP = 0
    FP = 0
    
    if TP == 0 and FP == 0: #If there are no True Positives and no False Positives.
        p = float("Nan")
        print("No True Positives or False Positives.")
    else:
        for i in range(N):
            if y_true[i] == True and y_pred[i] == True: #True Positive
                TP = TP + 1
            elif y_true[i] == False and y_pred[i] == True: #False Positive
                FP = FP + 1
        p = (TP)/(TP + FP)
    return p

##----------------------------------------------------------------------------------------

def MyRecall(y_true, y_pred):
    # Using the formula a = (TP)/(TP+FN)
    
    #Total number of datapoints
    N = len(y_true)
    
    #Initialize True Positive and False Negative
    TP = 0
    FN = 0
    
    if TP == 0 and FN == 0: #If there are no True Positives and no False Negatives
        r = float("NaN")
        print("No True Positives or False Negatives.")
    else:
        for i in range(N):
            if y_true[i] == True and y_pred[i] == True: #True Positive
                TP = TP + 1
            elif y_true[i] == True and y_pred[i] == False: #False Negative
                FN = FN+ 1
        r = (TP)/(TP + FN)
    return r

##----------------------------------------------------------------------------------------

def MyF1(y_true, y_pred):
    r = MyRecall(y_true, y_pred)
    p = MyPrecision(y_true, y_pred)
    
    F1 = (2*p*r)/(p + r)
    
    return F1
````

With the various score functions, they can now be compared to the built in sklearn scores:
